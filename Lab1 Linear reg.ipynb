```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train=np.array([2104,1416,1534,852])#size in sq feet
y_train=np.array([400,232,315,178])#price in $
# x_train=np.array([1.0,2.0])#size in sq feet
# y_train=np.array([300,500])#price in $
print(f"The value of X_train:{x_train}")
print(f"The value of y_train:{y_train}")
```

```python
#m is the number of training example
print(f"x_train.shape:{x_train.shape}")
m=y_train.shape[0]
print(f"Num of training example is:{m}")
#another method to find the num of training example
m1=len(x_train)
print(m)
```

```python
#plot the values
plt.scatter(x_train,y_train,c='r',marker='*',label='data point')
plt.title("house pricing")
plt.xlabel("size in  sq feet")
plt.ylabel("price in( 1000s dollar)")
plt.legend()
plt.show()
```

```python
b=120
w=0.1
print(f"b={b}")
print(f"w={w}")
```

#Model linear regression

```python
#comput model output
def compute_model_output(x,w,b):
    m=x.shape[0]
    f_wb=np.zeros(m)
    print(f_wb)
    for i in range(m):
        f_wb[i]=w*x[i]+b
    '''for single value prediction'''
    # f_wb=w*x+b
    return f_wb
```

```python
tmp_f_wb=compute_model_output(x_train,w,b)
print(tmp_f_wb)
plt.plot(x_train,tmp_f_wb,c='r',label='our prediction')
plt.scatter(x_train,y_train,c='b',marker='x',label='actual value')
plt.title("House Pricing")
plt.xlabel("Price in (1000s dollar)")
plt.xlabel("Size in 1000s sqft")
plt.legend()
plt.show()
```
